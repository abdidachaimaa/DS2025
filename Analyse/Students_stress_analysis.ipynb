### Analyse de R√©gression Lin√©aire

Nous allons visualiser la relation entre la charge d'√©tude et les niveaux de stress √† l'aide d'une r√©gression lin√©aire.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='how would you rate your study load?', y='How would you rate your stress levels?', data=student_df, line_kws={'color': 'red'})
plt.title('R√©gression Lin√©aire : Charge d\'√©tude vs Niveaux de Stress')
plt.xlabel('Charge d\'√©tude')
plt.ylabel('Niveaux de Stress')
plt.grid(True)
plt.show()
```

### Analyse de R√©gression Logistique

Pour la r√©gression logistique, nous allons d'abord cr√©er une variable binaire pour les niveaux de stress (par exemple, 'stress √©lev√©' si le niveau de stress est sup√©rieur √† 3, sinon 'stress faible'). Ensuite, nous analyserons la relation entre la qualit√© du sommeil et cette nouvelle variable de stress binaire.

```python
# Cr√©er une variable binaire pour le stress √©lev√©
student_df['High_Stress'] = (student_df['How would you rate your stress levels?'] > 3).astype(int)

plt.figure(figsize=(10, 6))
sns.regplot(x='Kindly Rate your Sleep Quality üò¥', y='High_Stress', data=student_df, logistic=True, ci=None, line_kws={'color': 'red'})
plt.title('R√©gression Logistique : Qualit√© du Sommeil vs Stress √âlev√©')
plt.xlabel('Qualit√© du Sommeil')
plt.ylabel('Stress √âlev√© (1=Oui, 0=Non)')
plt.grid(True)
plt.show()
```
